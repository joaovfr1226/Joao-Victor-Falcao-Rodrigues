```python
import sqlite3
import pandas as pd

# Caminho do banco de dados
db_path = r'C:\Users\jvfal\OneDrive\Área de Trabalho\Teste Analista Jr\data_analytics_jr_test\Questão 1\test_analytics.db'

# Conectar ao banco de dados SQLite
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Função para executar a consulta SQL e salvar os resultados em CSV
def consultar_e_salvar_csv(query, nome_arquivo):
    # Executar a consulta SQL
    df = pd.read_sql(query, conn)
    
    # Salvar os resultados em CSV
    df.to_csv(nome_arquivo, index=False)
    print(f'Arquivo {nome_arquivo} salvo com sucesso!')

# 1. Alunos por sexo
query_sexo = """
SELECT 
    alunos.SEXO, 
    COUNT(*) AS num_alunos
FROM 
    alunos
GROUP BY 
    alunos.SEXO
ORDER BY 
    num_alunos DESC;
"""
consultar_e_salvar_csv(query_sexo, 'alunos_por_sexo.csv')

# 2. Alunos por turno (análise por turno de aula)
query_turno = """
SELECT 
    alunos.TURNO, 
    COUNT(*) AS num_alunos
FROM 
    alunos
GROUP BY 
    alunos.TURNO
ORDER BY 
    num_alunos DESC;
"""
consultar_e_salvar_csv(query_turno, 'alunos_por_turno.csv')

# 3. Alunos por faixa etária
query_faixa_etaria = """
SELECT 
    alunos.IDADE, 
    COUNT(*) AS num_alunos
FROM 
    alunos
GROUP BY 
    alunos.IDADE
ORDER BY 
    alunos.IDADE;
"""
consultar_e_salvar_csv(query_faixa_etaria, 'alunos_por_faixa_etaria.csv')

# 4. Alunos por tipo de escola (ex: pública, privada, etc.)
query_tipo_escola = """
SELECT 
    escolas.TIPOESC, 
    COUNT(alunos.CODESC) AS num_alunos
FROM 
    alunos
JOIN 
    escolas ON alunos.CODESC = escolas.CODESC
GROUP BY 
    escolas.TIPOESC
ORDER BY 
    num_alunos DESC;
"""
consultar_e_salvar_csv(query_tipo_escola, 'alunos_por_tipo_de_escola.csv')

# 5. Alunos por DRE (Diretoria Regional de Educação)
query_dre = """
SELECT 
    escolas.DRE, 
    COUNT(alunos.CODESC) AS num_alunos
FROM 
    alunos
JOIN 
    escolas ON alunos.CODESC = escolas.CODESC
GROUP BY 
    escolas.DRE
ORDER BY 
    num_alunos DESC;
"""
consultar_e_salvar_csv(query_dre, 'alunos_por_dre.csv')

# 6. Alunos por tipo de modalidade de ensino (ex: EJA, Ensino Fundamental, etc.)
query_modalidade = """
SELECT 
    alunos.MODAL, 
    COUNT(*) AS num_alunos
FROM 
    alunos
GROUP BY 
    alunos.MODAL
ORDER BY 
    num_alunos DESC;
"""
consultar_e_salvar_csv(query_modalidade, 'alunos_por_modalidade.csv')

# 7. Alunos com necessidades educacionais especiais por sexo
query_nee_sexo = """
SELECT 
    alunos.SEXO, 
    alunos.NEE, 
    COUNT(*) AS num_alunos
FROM 
    alunos
WHERE 
    alunos.NEE IS NOT NULL
GROUP BY 
    alunos.SEXO, alunos.NEE
ORDER BY 
    num_alunos DESC;
"""
consultar_e_salvar_csv(query_nee_sexo, 'alunos_com_nee_por_sexo.csv')

# 8. Alunos por bairro e tipo de escola
query_bairro_tipo = """
SELECT 
    escolas.BAIRRO, 
    escolas.TIPOESC, 
    COUNT(alunos.CODESC) AS num_alunos
FROM 
    alunos
JOIN 
    escolas ON alunos.CODESC = escolas.CODESC
GROUP BY 
    escolas.BAIRRO, escolas.TIPOESC
ORDER BY 
    num_alunos DESC;
"""
consultar_e_salvar_csv(query_bairro_tipo, 'alunos_por_bairro_tipo_de_escola.csv')

# 9. Alunos por faixa etária e turno
query_faixa_etaria_turno = """
SELECT 
    alunos.IDADE, 
    alunos.TURNO, 
    COUNT(*) AS num_alunos
FROM 
    alunos
GROUP BY 
    alunos.IDADE, alunos.TURNO
ORDER BY 
    alunos.IDADE, alunos.TURNO;
"""
consultar_e_salvar_csv(query_faixa_etaria_turno, 'alunos_por_faixa_etaria_turno.csv')

# 10. Alunos por bairro e DRE
query_bairro_dre = """
SELECT 
    escolas.BAIRRO, 
    escolas.DRE, 
    COUNT(alunos.CODESC) AS num_alunos
FROM 
    alunos
JOIN 
    escolas ON alunos.CODESC = escolas.CODESC
GROUP BY 
    escolas.BAIRRO, escolas.DRE
ORDER BY 
    num_alunos DESC;
"""
consultar_e_salvar_csv(query_bairro_dre, 'alunos_por_bairro_dre.csv')

# 11. Alunos com necessidades educacionais especiais 
query_nee_dre = """
SELECT 
    escolas.DRE, 
    alunos.NEE, 
    COUNT(*) AS num_alunos
FROM 
    alunos
JOIN 
    escolas ON alunos.CODESC = escolas.CODESC
WHERE 
    alunos.NEE IS NOT NULL
GROUP BY 
    escolas.DRE, alunos.NEE
ORDER BY 
    num_alunos DESC;
"""
consultar_e_salvar_csv(query_nee_dre, 'alunos_com_nee_por_dre.csv')

# Fecha a conexão com o banco de dados
conn.close()


```
